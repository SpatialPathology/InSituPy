```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Function to generate random blobs
def generate_random_blobs(shape, num_blobs, min_radius, max_radius):
    image = np.zeros(shape, dtype=np.uint8)
    for _ in range(num_blobs):
        x = np.random.randint(0, shape[1])
        y = np.random.randint(0, shape[0])
        radius = np.random.randint(min_radius, max_radius)
        color = np.random.randint(100, 256)
        cv2.circle(image, (x, y), radius, color, -1)
    return image

# Generate random example images with blobs
image1 = generate_random_blobs((512, 512), 20, 10, 50)
image2 = generate_random_blobs((512, 512), 20, 10, 50)
image3 = generate_random_blobs((512, 512), 20, 10, 50)
image4 = generate_random_blobs((512, 512), 20, 10, 50)
image5 = generate_random_blobs((512, 512), 20, 10, 50)
image6 = generate_random_blobs((512, 512), 20, 10, 50)

# Function to create a colored image with different intensities of a single color
def create_colored_image(image, color):
    colored_image = np.zeros((image.shape[0], image.shape[1], 3), dtype=np.uint8)
    for i in range(3):
        colored_image[:, :, i] = image * color[i]
    return colored_image

# Define colors (BGR format)
red = [0, 0, 1]
green = [0, 1, 0]
blue = [1, 0, 0]
yellow = [0, 1, 1]
cyan = [1, 1, 0]
magenta = [1, 0, 1]

# Apply colors
image1_color = create_colored_image(image1, red)
image2_color = create_colored_image(image2, green)
image3_color = create_colored_image(image3, blue)
image4_color = create_colored_image(image4, yellow)
image5_color = create_colored_image(image5, cyan)
image6_color = create_colored_image(image6, magenta)

# Blend the images
blended_image = cv2.addWeighted(image1_color, 1/6.0, image2_color, 1/6.0, 0)
blended_image = cv2.addWeighted(blended_image, 1.0, image3_color, 1/6.0, 0)
blended_image = cv2.addWeighted(blended_image, 1.0, image4_color, 1/6.0, 0)
blended_image = cv2.addWeighted(blended_image, 1.0, image5_color, 1/6.0, 0)
blended_image = cv2.addWeighted(blended_image, 1.0, image6_color, 1/6.0, 0)

# Plot the images
fig, axes = plt.subplots(2, 4, figsize=(20, 10))
axes[0][0].imshow(cv2.cvtColor(image1_color, cv2.COLOR_BGR2RGB))
axes[0][0].set_title('Red Channel')
axes[0][1].imshow(cv2.cvtColor(image2_color, cv2.COLOR_BGR2RGB))
axes[0][1].set_title('Green Channel')
axes[0][2].imshow(cv2.cvtColor(image3_color, cv2.COLOR_BGR2RGB))
axes[0][2].set_title('Blue Channel')
axes[0][3].imshow(cv2.cvtColor(image4_color, cv2.COLOR_BGR2RGB))
axes[0][3].set_title('Yellow Channel')
axes[1][0].imshow(cv2.cvtColor(image5_color, cv2.COLOR_BGR2RGB))
axes[1][0].set_title('Cyan Channel')
axes[1][1].imshow(cv2.cvtColor(image6_color, cv2.COLOR_BGR2RGB))
axes[1][1].set_title('Magenta Channel')
axes[1][3].imshow(cv2.cvtColor(blended_image, cv2.COLOR_BGR2RGB))
axes[1][3].set_title('Blended Image')

# Hide the empty subplot
axes[1][2].axis('off')

plt.tight_layout()
plt.show()
```

## With points

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate random points with x, y values and six different intensity values
num_points = 100
x = np.random.rand(num_points) * 100
y = np.random.rand(num_points) * 100
intensity_values = np.random.rand(num_points, 6)

# Define colors (R, G, B) for Reds, Greens, Blues, Yellows, Cyans, and Magentas
colors = np.array([
    [1, 0, 0],  # Red
    [0, 1, 0],  # Green
    [0, 0, 1],  # Blue
    [1, 1, 0],  # Yellow (Red + Green)
    [0, 1, 1],  # Cyan (Green + Blue)
    [1, 0, 1]   # Magenta (Red + Blue)
])

# Plot the points with individual colors before merging
fig, axes = plt.subplots(2, 4, figsize=(20, 10))
axes = axes.ravel()
for i in range(6):
    ax = axes[i]
    ax.scatter(x, y, c=intensity_values[:, i][:, None] * colors[i], s=100)
    ax.set_title(f'Color {i+1}')
    ax.set_xlabel('X')
    ax.set_ylabel('Y')

# Blend the colors per point based on intensity values
blended_colors = np.dot(intensity_values, colors)

# Normalize the blended colors to be in the range [0, 1]
blended_colors = blended_colors / blended_colors.max(axis=1)[:, None]

# Plot the points using a scatter plot with blended colors
axes[-1].scatter(x, y, c=blended_colors, s=100)
axes[-1].set_title('Blended Colors')
axes[-1].set_xlabel('X')
axes[-1].set_ylabel('Y')

plt.tight_layout()
plt.show()
```
