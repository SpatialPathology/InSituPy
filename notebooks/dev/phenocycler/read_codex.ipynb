```python
# Calculate the center point for X
df['XCenter'] = (df['XMin'] + df['XMax']) / 2

# Calculate the center point for Y
df['YCenter'] = (df['YMin'] + df['YMax']) / 2
```

```python
df
```

```python
cyto_intensity = df[[elem for elem in df.columns if "Cytoplasm Intensity" in elem]]
cell_intensity = df[[elem for elem in df.columns if "Cell Intensity" in elem]]
cell_intensity = cell_intensity[[elem for elem in cell_intensity.columns if not "DAPI" in elem]]
coords = df[["XCenter", "YCenter"]].values
obs_names = df["Object Id"].values
var_names = [elem.split(" ")[0] for elem in cyto_intensity.columns]
```

```python
adata = ad.AnnData(X=cell_intensity.values)
adata.obs.index = obs_names
adata.var_names = var_names
adata.obsm["spatial"] = coords
```
