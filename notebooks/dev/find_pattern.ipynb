```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
from sklearn.metrics import mean_squared_error

# Example data
np.random.seed(0)
x = np.linspace(0, 10, 100)
y_linear_corr = 2 * x + 1 + np.random.normal(0, 1, 100)
y_linear_anti_corr = -2 * x + 20 + np.random.normal(0, 1, 100)
y_peak_center = -1 * (x - 5)**2 + 25 + np.random.normal(0, 1, 100)

# Define the models
def linear_corr(x, a, b):
    return a * x + b

def linear_anti_corr(x, a, b):
    return a * x + b

def peak_center(x, a, b, c):
    return a * (x - b)**2 + c

# Fit the models to the data
params_linear_corr, _ = curve_fit(linear_corr, x, y_linear_corr)
params_linear_anti_corr, _ = curve_fit(linear_anti_corr, x, y_linear_anti_corr)
params_peak_center, _ = curve_fit(peak_center, x, y_peak_center)

# Predict the values using the fitted models
y_pred_linear_corr = linear_corr(x, *params_linear_corr)
y_pred_linear_anti_corr = linear_anti_corr(x, *params_linear_anti_corr)
y_pred_peak_center = peak_center(x, *params_peak_center)

# Calculate the mean squared error for each model
mse_linear_corr = mean_squared_error(y_linear_corr, y_pred_linear_corr)
mse_linear_anti_corr = mean_squared_error(y_linear_anti_corr, y_pred_linear_anti_corr)
mse_peak_center = mean_squared_error(y_peak_center, y_pred_peak_center)

# Print the mean squared errors
print(f"Mean Squared Error for Linear Correlated Model: {mse_linear_corr}")
print(f"Mean Squared Error for Linear Anti-Correlated Model: {mse_linear_anti_corr}")
print(f"Mean Squared Error for Peak Center Model: {mse_peak_center}")

# Plot the data and the fitted models
plt.figure(figsize=(18, 6))

plt.subplot(1, 3, 1)
plt.scatter(x, y_linear_corr, label='Data')
plt.plot(x, y_pred_linear_corr, color='red', label='Fitted Model')
plt.title('Linear Correlated Model')
plt.legend()

plt.subplot(1, 3, 2)
plt.scatter(x, y_linear_anti_corr, label='Data')
plt.plot(x, y_pred_linear_anti_corr, color='red', label='Fitted Model')
plt.title('Linear Anti-Correlated Model')
plt.legend()

plt.subplot(1, 3, 3)
plt.scatter(x, y_peak_center, label='Data')
plt.plot(x, y_pred_peak_center, color='red', label='Fitted Model')
plt.title('Peak Center Model')
plt.legend()

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
from sklearn.metrics import mean_squared_error

# Example data
np.random.seed(0)
x = np.linspace(0, 10, 100)
y_linear_corr = 2 * x + 1 + np.random.normal(0, 1, 100)
y_linear_anti_corr = -2 * x + 20 + np.random.normal(0, 1, 100)
y_peak_center = -1 * (x - 5)**2 + 25 + np.random.normal(0, 1, 100)

# Define the models
def linear_corr(x, a, b):
    return a * x + b

def linear_anti_corr(x, a, b):
    return a * x + b

def peak_center(x, a, b, c):
    return a * (x - b)**2 + c

# Fit the models to the data
params_linear_corr, _ = curve_fit(linear_corr, x, y_linear_corr)
params_linear_anti_corr, _ = curve_fit(linear_anti_corr, x, y_linear_anti_corr)
params_peak_center, _ = curve_fit(peak_center, x, y_peak_center)

# Predict the values using the fitted models
y_pred_linear_corr = linear_corr(x, *params_linear_corr)
y_pred_linear_anti_corr = linear_anti_corr(x, *params_linear_anti_corr)
y_pred_peak_center = peak_center(x, *params_peak_center)

# Calculate the mean squared error for each model
mse_linear_corr = mean_squared_error(y_linear_corr, y_pred_linear_corr)
mse_linear_anti_corr = mean_squared_error(y_linear_anti_corr, y_pred_linear_anti_corr)
mse_peak_center = mean_squared_error(y_peak_center, y_pred_peak_center)

# Calculate the root mean squared error for each model
rmse_linear_corr = np.sqrt(mse_linear_corr)
rmse_linear_anti_corr = np.sqrt(mse_linear_anti_corr)
rmse_peak_center = np.sqrt(mse_peak_center)

# Print the mean squared errors and root mean squared errors
print(f"Mean Squared Error for Linear Correlated Model: {mse_linear_corr}")
print(f"Root Mean Squared Error for Linear Correlated Model: {rmse_linear_corr}")
print(f"Mean Squared Error for Linear Anti-Correlated Model: {mse_linear_anti_corr}")
print(f"Root Mean Squared Error for Linear Anti-Correlated Model: {rmse_linear_anti_corr}")
print(f"Mean Squared Error for Peak Center Model: {mse_peak_center}")
print(f"Root Mean Squared Error for Peak Center Model: {rmse_peak_center}")

# Plot the data and the fitted models
plt.figure(figsize=(18, 6))

plt.subplot(1, 3, 1)
plt.scatter(x, y_linear_corr, label='Data')
plt.plot(x, y_pred_linear_corr, color='red', label='Fitted Model')
plt.title('Linear Correlated Model')
plt.legend()

plt.subplot(1, 3, 2)
plt.scatter(x, y_linear_anti_corr, label='Data')
plt.plot(x, y_pred_linear_anti_corr, color='red', label='Fitted Model')
plt.title('Linear Anti-Correlated Model')
plt.legend()

plt.subplot(1, 3, 3)
plt.scatter(x, y_peak_center, label='Data')
plt.plot(x, y_pred_peak_center, color='red', label='Fitted Model')
plt.title('Peak Center Model')
plt.legend()

plt.show()
```
